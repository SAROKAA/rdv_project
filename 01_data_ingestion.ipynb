```python
import pandas as pd
import duckdb
import os
```

```python
months = [
    "2025-01",
    "2025-02",
    "2025-03",
    "2025-04",
    "2025-05",
    "2025-06"
]

base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data/"
```

```python
raw_path = "/content/drive/MyDrive/RDV_Project/raw_data"

os.makedirs(raw_path, exist_ok=True)

print("Folder siap:", raw_path)
```

```python
import urllib.request

for month in months:
    file_name = f"yellow_tripdata_{month}.parquet"
    url = base_url + file_name

    save_path = os.path.join(raw_path, file_name)

    if not os.path.exists(save_path):
        print(f"Downloading {file_name}...")
        urllib.request.urlretrieve(url, save_path)
        print(f"Saved: {save_path}")
    else:
        print(f"{file_name} already exists")
```

```python
sample_file = os.path.join(
    raw_path,
    "yellow_tripdata_2025-01.parquet"
)

df = pd.read_parquet(sample_file)

df.head()
```

```python
selected_columns = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime",
    "PULocationID",
    "DOLocationID",
    "trip_distance",
    "fare_amount"
]
```

```python
all_data = []

for month in months:
    file_path = os.path.join(
        raw_path,
        f"yellow_tripdata_{month}.parquet"
    )

    df = pd.read_parquet(
        file_path,
        columns=selected_columns
    )

    df["month"] = month

    all_data.append(df)

combined_df = pd.concat(
    all_data,
    ignore_index=True
)

print(combined_df.shape)
combined_df.head()
```

```python
output_path = "/content/drive/MyDrive/RDV_Project/processed_data"

os.makedirs(output_path, exist_ok=True)

combined_df.to_parquet(
    f"{output_path}/taxi_raw_ingested.parquet",
    index=False
)

print("Raw ingested data saved")
```
