```python
app_code = '''
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

st.set_page_config(
    page_title="NYC Yellow Taxi Dashboard",
    layout="wide"
)

# ======================
# LOAD DATA
# ======================

BASE_PATH = (
    "/content/drive/MyDrive/"
    "RDV_Project/processed_data/"
)

taxi_data = pd.read_parquet(
    BASE_PATH + "taxi_weather.parquet"
)

trip_by_hour = pd.read_parquet(
    BASE_PATH + "agg_trip_by_hour.parquet"
)

trip_characteristics = pd.read_parquet(
    BASE_PATH + "agg_trip_characteristics.parquet"
)

top_pickup_zone = pd.read_parquet(
    BASE_PATH + "agg_top_pickup_zone.parquet"
)

weather_effect = pd.read_parquet(
    BASE_PATH + "agg_weather_effect.parquet"
)

prediction = pd.read_parquet(
    BASE_PATH + "trip_prediction.parquet"
)

# ======================
# FEATURE ENGINEERING
# ======================

taxi_data["pickup_date"] = pd.to_datetime(
    taxi_data["pickup_date"]
)

taxi_data["day_name"] = (
    taxi_data["pickup_date"]
    .dt.day_name()
)

# ======================
# HEADER
# ======================

st.markdown("""
<h1 style='color:#0d2b5c;'>
🚖 NYC Yellow Taxi – Analisis Pola Mobilitas
</h1>

<p style='font-size:18px;'>
Pola perjalanan berdasarkan waktu,
lokasi, dan cuaca
(Jan – Jun 2025)
</p>
""", unsafe_allow_html=True)

st.divider()

# ======================
# KPI CARDS
# ======================

col1, col2, col3, col4, col5 = st.columns(5)

col1.metric(
    "Total Perjalanan",
    f"{len(taxi_data):,}"
)

col2.metric(
    "Rata-rata Jarak",
    f"{taxi_data['trip_distance'].mean():.2f} km"
)

col3.metric(
    "Rata-rata Durasi",
    f"{taxi_data['trip_duration'].mean():.2f} menit"
)

col4.metric(
    "Rata-rata Fare",
    f"${taxi_data['fare_amount'].mean():.2f}"
)

dominant_weather = (
    taxi_data[
        "weather_condition"
    ]
    .mode()[0]
)

col5.metric(
    "Cuaca Dominan",
    dominant_weather
)

st.divider()

# ======================
# ROW 1
# ======================

left, middle, right = st.columns([1.1,1,1])

# ----------------------
# TOP PICKUP
# ----------------------

with left:

    st.subheader(
        "Zona Pickup Tertinggi"
    )

    fig_pickup = px.bar(
        top_pickup_zone.head(15),
        x="PULocationID",
        y="total_pickup",
        color="total_pickup"
    )

    st.plotly_chart(
        fig_pickup,
        use_container_width=True
    )

# ----------------------
# TRIP BY HOUR
# ----------------------

with middle:

    st.subheader(
        "Distribusi Perjalanan per Jam"
    )

    fig_hour = px.bar(
        trip_by_hour,
        x="pickup_hour",
        y="total_trip"
    )

    st.plotly_chart(
        fig_hour,
        use_container_width=True
    )

# ----------------------
# WEATHER
# ----------------------

with right:

    st.subheader(
        "Pengaruh Cuaca"
    )

    fig_weather = px.bar(
        weather_effect,
        x="weather_condition",
        y="total_trip",
        color="weather_condition"
    )

    st.plotly_chart(
        fig_weather,
        use_container_width=True
    )

# ======================
# ROW 2
# ======================

left2, middle2, right2 = st.columns([1,1,1])

# ----------------------
# DAY OF WEEK
# ----------------------

with left2:

    st.subheader(
        "Jumlah Perjalanan per Hari"
    )

    day_order = [
        "Monday",
        "Tuesday",
        "Wednesday",
        "Thursday",
        "Friday",
        "Saturday",
        "Sunday"
    ]

    day_trip = (
        taxi_data.groupby(
            "day_name"
        )
        .size()
        .reset_index(
            name="total_trip"
        )
    )

    day_trip["day_name"] = pd.Categorical(
        day_trip["day_name"],
        categories=day_order,
        ordered=True
    )

    day_trip = day_trip.sort_values(
        "day_name"
    )

    fig_day = px.bar(
        day_trip,
        x="total_trip",
        y="day_name",
        orientation="h"
    )

    st.plotly_chart(
        fig_day,
        use_container_width=True
    )

# ----------------------
# DISTANCE
# ----------------------

with middle2:

    st.subheader(
        "Distribusi Jarak"
    )

    distance_data = taxi_data[
        taxi_data["trip_distance"].between(
            0, 20
        )
    ]

    fig_distance = px.histogram(
        distance_data,
        x="trip_distance",
        nbins=50,
        title="Distribusi Trip Distance"
    )

    fig_distance.update_layout(
        xaxis_title="Jarak Perjalanan (km)",
        yaxis_title="Jumlah Trip"
    )

    st.plotly_chart(
        fig_distance,
        use_container_width=True
    )

# ----------------------
# DURATION
# ----------------------

with right2:

    st.subheader(
        "Distribusi Durasi"
    )

    duration_data = taxi_data[
        taxi_data["trip_duration"].between(
            0, 120
        )
    ]

    fig_duration = px.histogram(
        duration_data,
        x="trip_duration",
        nbins=50,
        title="Distribusi Durasi"
    )

    fig_duration.update_layout(
        xaxis_title="Durasi (menit)",
        yaxis_title="Jumlah Trip"
    )

    st.plotly_chart(
        fig_duration,
        use_container_width=True
    )

# ======================
# TABLE
# ======================

st.subheader(
    "Ringkasan Karakteristik"
)

summary_table = taxi_data.groupby(
    "time_category"
).agg({
    "trip_distance":"mean",
    "trip_duration":"mean",
    "fare_amount":"mean"
}).round(2)

st.dataframe(
    summary_table,
    use_container_width=True
)

# ======================
# ML PREDICTION
# ======================

st.subheader(
    "Actual vs Predicted Trip"
)

fig_pred = px.scatter(
    prediction,
    x="actual_trip",
    y="predicted_trip",
    trendline="ols"
)

st.plotly_chart(
    fig_pred,
    use_container_width=True
)
'''
```

```python
app_code = '''
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(
    page_title="NYC Taxi Dashboard",
    layout="wide"
)

BASE_PATH = (
    "/content/drive/MyDrive/"
    "RDV_Project/processed_data/"
)

# =====================
# LOAD DATA
# =====================

taxi_data = pd.read_parquet(
    BASE_PATH + "dashboard_sample.parquet"
)

trip_by_hour = pd.read_parquet(
    BASE_PATH + "agg_trip_by_hour.parquet"
)

trip_characteristics = pd.read_parquet(
    BASE_PATH + "agg_trip_characteristics.parquet"
)

top_pickup_zone = pd.read_parquet(
    BASE_PATH + "agg_top_pickup_zone.parquet"
)

weather_effect = pd.read_parquet(
    BASE_PATH + "agg_weather_effect.parquet"
)

prediction = pd.read_parquet(
    BASE_PATH + "trip_prediction.parquet"
)

# =====================
# FEATURE ENGINEERING
# =====================

taxi_data["pickup_date"] = pd.to_datetime(
    taxi_data["pickup_date"]
)

taxi_data["day_name"] = (
    taxi_data["pickup_date"]
    .dt.day_name()
)

# =====================
# HEADER
# =====================

st.title(
    "🚖 NYC Yellow Taxi – Analisis Pola Mobilitas"
)

st.caption(
    "Pola perjalanan berdasarkan waktu, lokasi, dan cuaca (Jan–Jun 2025)"
)

# =====================
# KPI
# =====================

col1, col2, col3, col4 = st.columns(4)

col1.metric(
    "Total Trip",
    f"{len(taxi_data):,}"
)

col2.metric(
    "Avg Distance",
    f"{taxi_data['trip_distance'].mean():.2f} km"
)

col3.metric(
    "Avg Duration",
    f"{taxi_data['trip_duration'].mean():.2f} min"
)

col4.metric(
    "Avg Fare",
    f"${taxi_data['fare_amount'].mean():.2f}"
)

st.divider()

# =====================
# ROW 1
# =====================

c1, c2 = st.columns(2)

with c1:

    st.subheader(
        "Distribusi Perjalanan per Jam"
    )

    fig_hour = px.bar(
        trip_by_hour,
        x="pickup_hour",
        y="total_trip"
    )

    st.plotly_chart(
        fig_hour,
        use_container_width=True
    )

with c2:

    st.subheader(
        "Top Pickup Zone"
    )

    fig_pickup = px.bar(
        top_pickup_zone.head(10),
        x="PULocationID",
        y="total_pickup",
        color="total_pickup"
    )

    st.plotly_chart(
        fig_pickup,
        use_container_width=True
    )

# =====================
# ROW 2
# =====================

c3, c4 = st.columns(2)

with c3:

    st.subheader(
        "Distribusi Jarak"
    )

    fig_distance = px.histogram(
        taxi_data,
        x="trip_distance",
        nbins=40,
        range_x=[0,20]
    )

    st.plotly_chart(
        fig_distance,
        use_container_width=True
    )

with c4:

    st.subheader(
        "Distribusi Durasi"
    )

    fig_duration = px.histogram(
        taxi_data,
        x="trip_duration",
        nbins=40,
        range_x=[0,120]
    )

    st.plotly_chart(
        fig_duration,
        use_container_width=True
    )

# =====================
# WEATHER
# =====================

st.subheader(
    "Pengaruh Cuaca"
)

fig_weather = px.bar(
    weather_effect,
    x="weather_condition",
    y="total_trip",
    color="weather_condition"
)

st.plotly_chart(
    fig_weather,
    use_container_width=True
)

# =====================
# PICKUP MAP
# =====================

st.subheader(
    "NYC Pickup Heatmap"
)

import json

with open(
    BASE_PATH + "taxi_zones.geojson"
) as f:
    geojson = json.load(f)

pickup_map = (
    taxi_data.groupby(
        "PULocationID"
    )
    .size()
    .reset_index(
        name="total_trip"
    )
)

fig_map = px.choropleth_mapbox(
    pickup_map,
    geojson=geojson,
    locations="PULocationID",
    featureidkey="properties.location_id",
    color="total_trip",
    mapbox_style="carto-positron",
    zoom=9,
    center={
        "lat": 40.7128,
        "lon": -74.0060
    },
    opacity=0.7
)

fig_map.update_layout(
    margin=dict(
        l=0,
        r=0,
        t=0,
        b=0
    )
)

st.plotly_chart(
    fig_map,
    use_container_width=True
)
# =====================
# HEATMAP
# =====================

st.subheader(
    "Heatmap Perjalanan Hari × Jam"
)

heatmap_data = (
    taxi_data.groupby(
        ["day_name", "pickup_hour"]
    )
    .size()
    .reset_index(
        name="total_trip"
    )
)

day_order = [
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday"
]

heatmap_data["day_name"] = pd.Categorical(
    heatmap_data["day_name"],
    categories=day_order,
    ordered=True
)

heatmap_pivot = heatmap_data.pivot(
    index="day_name",
    columns="pickup_hour",
    values="total_trip"
)

fig_heatmap = px.imshow(
    heatmap_pivot,
    aspect="auto",
    labels=dict(
        x="Jam",
        y="Hari",
        color="Jumlah Trip"
    )
)

st.plotly_chart(
    fig_heatmap,
    use_container_width=True
)

# =====================
# BOXPLOT
# =====================

st.subheader(
    "Fare Distribution by Time Category"
)

fig_box = px.box(
    taxi_data,
    x="time_category",
    y="fare_amount",
    color="time_category"
)

fig_box.update_yaxes(
    range=[0,100]
)

st.plotly_chart(
    fig_box,
    use_container_width=True
)

# =====================
# CORRELATION
# =====================

st.subheader(
    "Correlation Heatmap"
)

corr_data = taxi_data[
    [
        "trip_distance",
        "trip_duration",
        "fare_amount",
        "pickup_hour"
    ]
].corr()

fig_corr = px.imshow(
    corr_data,
    text_auto=True,
    aspect="auto"
)

st.plotly_chart(
    fig_corr,
    use_container_width=True
)

# =====================
# ML
# =====================

st.subheader(
    "Actual vs Predicted Trip"
)

fig_pred = px.scatter(
    prediction,
    x="actual_trip",
    y="predicted_trip",
    trendline="ols"
)

st.plotly_chart(
    fig_pred,
    use_container_width=True
)
'''
```

```python
with open(app_file, "w") as f:
    f.write(app_code)

print("Dashboard updated")
```
